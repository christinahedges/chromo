```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ellipsoidal_variation(A, phase):
    return A * np.cos(4 * np.pi * phase)
```

```python
# Check shporer 2017 for limb darkening in EBs and effect on Amplitude of elipsoidal
```

```python
plt.plot(ellipsoidal_variation(-1e3, np.linspace(0, 1, 1000)))
```
